```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人であること")
        else:
            print(f"{self.name}は大人でないこと")

# Humanクラスのインスタンスを複数生成し、リストに追加
humans = [
    Human("Alice", 25),
    Human("Bob", 18),
    Human("Charlie", 30),
    Human("David", 15)
]

# リストの要素数分だけcheck_adultメソッドを呼び出す
for human in humans:
    human.check_adult()
```
