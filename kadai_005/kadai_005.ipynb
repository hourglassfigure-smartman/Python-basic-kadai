```python
# 台形の面積を計算するプログラム
# 上辺
upper = 10
# 下辺
lower = 20
# 高さ
height = 5
# 公式　：台形の面積 =（上辺＋下辺）× 高さ ÷ 2
result = (upper + lower) * height / 2
# 台形の面積の計算結果は、変数に代入して「◯◯cm²」となるように位を表示
print(f"{result}cm²")
```
