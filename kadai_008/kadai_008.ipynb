```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 範囲の指定がないため変数varに0～15までのランダムな整数を代入する
var = random.randint(1,15)

# 例として1から15までの数で試します
if var % 3 == 0 and var % 5 == 0:
    print("FizzBuzz")
elif var % 3 == 0:
    print("Fizz")
elif var % 5 == 0:
    print("Buzz")
else:
    print(var)
```
