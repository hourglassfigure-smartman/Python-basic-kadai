```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def printinfo(self):
        print(f"名前: {self.name}, 年齢: {self.age}")

# Humanクラスのインスタンスを作成
human_instance = Human("Alice", 30)

# printinfoメソッドを呼び出し
human_instance.printinfo()
```
