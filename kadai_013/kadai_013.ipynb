```python
def calculate_price_with_tax(item_price, tax_rate_percentage):
    """
    商品の金額と消費税率（パーセンテージ）を受け取り、
    消費税込みの合計金額を計算して返します。

    Args:
        item_price (float or int): 商品の金額。
        tax_rate_percentage (float or int): 消費税率（例: 10%の場合は10）。

    Returns:
        float: 消費税込みの合計金額。
    """
    tax_amount = item_price * (tax_rate_percentage / 100)
    total_price = item_price + tax_amount
    return total_price

# 例:
# 第1引数に商品の金額「110」、第2引数に消費税（10%）の「10」を設定
item_price = 110
tax_rate = 10

result = calculate_price_with_tax(item_price, tax_rate)
print(f"商品金額: {item_price}円")
print(f"消費税率: {tax_rate}%")
print(f"消費税込みの合計金額: {result}円")
```
